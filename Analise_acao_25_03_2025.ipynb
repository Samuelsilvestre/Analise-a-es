```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

Carregando os dados

```python

carrefour_symbol = 'CRFB3.SA'
pao_de_acucar_symbol = 'PCAR3.SA'
cr = yf.download(carrefour_symbol, start='2020-01-01', end='2025-03-25')
pa = yf.download(pao_de_acucar_symbol, start='2020-01-01', end='2025-03-25')

colunas_pt = {
    'Open': 'Abertura',
    'High': 'Máxima',
    'Low': 'Mínima',
    'Close': 'Fechamento',
    'Adj Close': 'Fechamento Ajustado',
    'Volume': 'Volume'
}
cr.rename(columns=colunas_pt, inplace=True)
pa.rename(columns=colunas_pt, inplace=True)
```

<p>Valor de fechamento de 2020 a 2025</p>

```python

plt.figure(figsize=(15, 6))
plt.plot(pa['Fechamento'], label='Pão de Açúcar', linewidth=2, alpha=0.8, color='blue')
plt.plot(cr['Fechamento'], label='Carrefour', linewidth=2, alpha=0.8, color='green')

plt.title('Fechamento das Ações ao Longo do Tempo', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Fechamento (R$)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5, linestyle='--')
plt.tight_layout()

plt.show()
```

<h2>Simulando a variação dos investimentos</h2>

<p>Variação percentual de 2 Mil ações</p>

```python
cr['Investimento'] = cr['Fechamento'] * 2000
pa['Investimento'] = pa['Fechamento'] * 2000
pa['%Investimento'] = pa['Investimento'].pct_change()
cr['%Investimento'] = cr['Investimento'].pct_change()
```

```python
pa_tm = pa['%Investimento'].resample('3M').mean()
cr_tm = cr['%Investimento'].resample('3M').mean()


plt.figure(figsize=(15, 6))
plt.plot(pa_tm, label='Pão de Açúcar', linewidth=2, alpha=0.8, color='blue')
plt.plot(cr_tm, label='Carrefour', linewidth=2, alpha=0.8, color='green')

plt.title('Média da Variação dos Investimentos - Trimestre', fontsize=16)
plt.xlabel('Data', fontsize=12)  # Rótulo do eixo X
plt.ylabel('Média pct', fontsize=12)  # Rótulo do eixo Y
plt.legend(fontsize=12)
plt.grid(alpha=0.5, linestyle='--')
plt.tight_layout()

plt.show();
```
