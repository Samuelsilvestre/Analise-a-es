<h2>Analise dos dados de negociação do IBOVESPA, Carrefour, Locaweb</h3>

<h3>Pacotes python</h3>

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h3>Baixar os dados dos últimos 6 meses do Ibovespa</h3>

```python

ibvsp = yf.download("^BVSP", period="6mo")

ibvsp.rename(columns={
    "Open": "Abertura",
    "High": "Alta",
    "Low": "Baixa",
    "Close": "Fechamento",
    "Adj Close": "Fechamento Ajustado",
    "Volume": "Volume"
}, inplace=True)

ibvsp
```

<h3>Média movel do indice IBVSP</H3>

```python
media_movel_15 = ibvsp["MediaMovel"] = ibvsp["Fechamento"].rolling(window=15).mean()
media_movel_30 = ibvsp["MediaMovel"] = ibvsp["Fechamento"].rolling(window=30).mean()

```

```python
plt.figure(figsize=(12,6))
plt.plot(media_movel_15, label="Média Movel 15 Dias", color="blue")
plt.plot(media_movel_30, label="Média Movel 30 Dias", color="green")
plt.title("Média Movel do Indice IBVS")
plt.xlabel("Data")
plt.ylabel("Fechamento")
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show();
```

<h1>Analise das ações do Carrefour</h1>

```python
carrefour = yf.download("CRFB3.SA", period="6mo")
carrefour.rename(columns={
    "Open": "Abertura",
    "High": "Alta",
    "Low": "Baixa",
    "Close": "Fechamento",
    "Adj Close": "Fechamento Ajustado",
    "Volume": "Volume"
}, inplace=True)

carrefour
```

<h3>Média Movel de 30 Dias</h3>

```python
media_movel_30 = carrefour["MediaMovel"] = carrefour["Fechamento"].rolling(window=40).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(media_movel_30, label="Média Movel", color="blue")
plt.title("Média Movel 30 Dias do Carrefour")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show();
```

<h3>Alta & Baixa</h3>

```python
plt.figure(figsize=(12,6))
plt.plot(carrefour.index, carrefour["Alta"], label="Alta", color="blue")
plt.plot(carrefour.index, carrefour["Baixa"], label="Baixa", color="red")
plt.title("Alta & Baixa das Ação do Carrefour")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show();
```

<h3>Analise da LocaWeb</h3>

```python
lw = yf.download("LWSA3.SA", period="6mo")
lw.rename(columns={
    "Open": "Abertura",
    "High": "Alta",
    "Low": "Baixa",
    "Close": "Fechamento",
    "Adj Close": "Fechamento Ajustado",
    "Volume": "Volume"
}, inplace=True)

lw
```

<h3>Média Movel de 7 Dias</h3>

```python
media_movel_7 = lw["MediaMovel"] = lw["Fechamento"].rolling(window=7).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(media_movel_7, label="Média Movel", color="blue")
plt.title("Média Movel 7 Dias da LocaWeb")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show();
```

<h3>Valor de Fechamento</h3>

```python
plt.figure(figsize=(12,6))
plt.plot(lw["Fechamento"], label="Fechamento", color="green")
plt.title("Valor de Fechamento da LocaWeb")
plt.xlabel("Data")
plt.ylabel("Valor")
plt.grid(alpha=0.5)
plt.legend(fontsize=12)
plt.show();
```
